```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import nitime
```

```python
filename = 'OpenBCI-RAW-2019-10-25_17-13-59.txt'
df = pd.read_csv(filename)
rel_cols = ['e{}'.format(i) for i in range(1,9)]
df = df[rel_cols]
```

```python
#elektrody co nas interesuja
n_samples = 20
sampling_freq = 250
e3 = df['e3'].to_numpy()[:n_samples*sampling_freq]
# e4 = df['e4']
e3 = np.reshape(e3, (n_samples,sampling_freq))
# e3 -= np.mean(e3, axis=0)
for row in range(n_samples):
    e3[row] -= np.mean(e3[row])
```

```python
from nitime.algorithms.autoregressive import AR_est_YW, AR_psd

n_observations = n_samples

a = np.zeros(n_observations)
sigmas = np.zeros(n_observations)

for i in range(n_observations):
    a[i], sigmas[i] = AR_est_YW(e3[i], 1)
# yw = AR_est_YW(e3], 1)
```

```python
x,y = AR_psd(a, sigmas)

plt.scatter(x,y)
```
